```python
import pandas as pd
from tqdm.notebook import tqdm
from math import log2
import plotly.graph_objects as go
```

```python
dg = pd.read_csv('genes.tsv',delimiter='\t')
dg = dg.rename(columns = {c:c.lower().replace(' ','_') for c in dg.columns})
dg = dg.rename(columns = {'approved_symbol':'symbol', 'approved_name':'name'})
dg = dg[['symbol','name']]
```

```python
### add info columns

dg['len'] = dg['symbol'].str.len()
dg['distinct'] = dg['symbol'].apply(lambda x: len(set(x)))

dg = dg[dg['len'].isin([3,4,5])].copy()
dg = dg[~dg['symbol'].str.contains('-')].copy()


dg['padded_symbol'] = dg['symbol'].apply(lambda s: s.ljust(5))
for i in range(0,5):
    dg[f'char{i+1}'] = dg['padded_symbol'].apply(lambda x: x[i])
```

```python
from collections import Counter

### frequency of characters based on the number of times a character appears in the corpus (counts repetition of characters in a single symbol)
def char_frequency(word_list):
    N = len(word_list)
    freq = Counter(''.join(word_list))
    #return {x:v/N for x,v in freq.most_common()}
    return freq

### frequency of characters based on how many symbols contain each character (ignores repetition of characters in a single symbol)
def symbol_frequency(word_list):
    N = len(word_list)
    word_list_distinct = [''.join(set(word)) for word in word_list]
    freq = Counter(''.join(word_list_distinct))
    #return {x:v/N for x,v in freq.most_common()}
    return freq

### based on how often each character appears in a given position 1-5 
def position_frequency(word_list):
    freq = {k:Counter() for k in range(5)}
    for word in word_list:
        for k in range(5):
            freq[k].update(word[k])
    return freq
```

```python
### initialize character freq dicts based on full corpus
freq_char = char_frequency(dg['padded_symbol'])
freq_symb = symbol_frequency(dg['padded_symbol'])
freq_pos = position_frequency(dg['padded_symbol'])
```

```python
def freq_score_sum(word, frequency=freq_char):
    return sum(frequency[x] for x in word)

def freq_score_sum_distinct(word, frequency=freq_char):
    return sum(frequency[x] for x in set(word))

def freq_score_sum_pos(word, frequency=freq_pos):
    return sum(frequency[k][x] for k,x in enumerate(word))
```

```python
dg['freq_char'] = dg['padded_symbol'].apply(lambda x:freq_score_sum(x,frequency=freq_char))
dg['freq_symbol'] = dg['padded_symbol'].apply(lambda x:freq_score_sum(x,frequency=freq_symb))
dg['freq_pos'] = dg['padded_symbol'].apply(lambda x:freq_score_sum_pos(x,frequency=freq_pos))
```

```python
dg = dg.sort_values('freq_pos', ascending=0)
```

```python
def overlap_matrix(words):
    da = pd.DataFrame(index=words)
    for w in tqdm(words):
        da[w] = list(map(lambda X: len(set(X)&set(w)), words))
    return da

words = list(dg['padded_symbol'])
da = overlap_matrix(words)
```

```python
class Words():
    def __init__(self,word_list):
        self.words = list(word_list)
        self.num_words = len(word_list)

        dw = pd.DataFrame()
        dw['word0'] = word_list
        dw['word'] = dw['word0'].apply(lambda w: w.ljust(5))
        dw = dw[~dw['word'].str.contains('-')].copy()
        for i in range(0,5):
            dw[f'char{i}'] = dw['word'].apply(lambda x: x[i])
        self.table = dw.copy()

    def update_list(self, guess, feedback):
        guess = guess.upper().ljust(5)
        dw = self.table.copy()
        fixed_chars = []
        good_chars = []
        bad_chars = set()
        for i,(x,f) in enumerate(zip(guess,feedback)):
            if f=='0':
                bad_chars.add(x)
            elif f=='1':
                good_chars.append(x)
                dw = dw[dw[f'char{i}'] != x]
            elif f=='2':
                fixed_chars.append(x)
                dw = dw[dw[f'char{i}'] == x]
        for x in bad_chars & set(good_chars+fixed_chars):
            bad_chars = bad_chars - {x}
            dw = dw[dw['word'].str.count(x)<= (good_chars+fixed_chars).count(x)]
        dw = dw[~dw['word'].str.contains('|'.join(bad_chars).upper())]
        dw = dw[dw['word'].apply(lambda s: all([x in s for x in good_chars]))==True]
        self.table = dw.copy() 
        self.words = sorted(self.table['word'])
        self.num_words = len(self.table)
```

```python
W = Words(dg['padded_symbol'])
W.update_list('rpa1','00000')
W.update_list('dctn2','00002')
W.update_list('olfm2','02002')
#W.update_list('dab1','22022')
#W.update_list('gad1','02122')
#W.update_list('vcx','22022')

dg_ = W.table.copy()
da_ = da.loc[W.words][W.words]

da_['overlap_mult'] = da_[W.words].sum(axis=1)
dg_['overlap_score'] = dg_['word'].apply(lambda x: da_.loc[x,'overlap_mult'])/(5*len(words))

dg_.sort_values('overlap_score',ascending=0)
```
